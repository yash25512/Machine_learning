# Hieraraichal clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# import iris dataset
iris=datasets.load_iris()
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
# Apply the pca because for dimensity reduction
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
## Agglomerative clustering
# to construct a dendogram

import scipy.cluster.hierarchy as sc
# plot the dendogram
plt.figure(figsize=(20,7))
plt.title("Dendograms")

# create dendogram
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=2,linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
## performance check silhouette score
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]

# we start at 2 clusters for silhouette cofficients
for k in range(2,11):
    agglo=AgglomerativeClustering(n_clusters=k,linkage='ward')
    agglo.fit(X_scaled)
    score=silhouette_score(X_scaled,agglo.labels_)
    silhouette_coefficients.append(score)
```

```python
# plotting silhoette score
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("num of clusters")
plt.ylabel("silhouette coeff")
plt.show()
```
