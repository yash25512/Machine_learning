# Linear regression with multiple variable

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv(r'C:\Users\namde\Downloads\homeprices (1).csv')
```

```python
df
```

```python
# data preprocessing 
import math
median_bedrooms=math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms=df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
from word2number import w2n
d.experience = d.experience.astype(str)
d.experience=d.experience.apply(w2n.word_to_num)
d
```

```python
import math
median_test_score = math.floor(d['test_score(out of 10)'].mean())
median_test_score
```

```python
d['test_score(out of 10)'] = d['test_score(out of 10)'].fillna(median_test_score)
d
```

```python
reg1=linear_model.LinearRegression()
reg1.fit(d[['experience','test_score(out of 10)','interview_score(out of 10)']],d['salary($)'])
```

```python
reg1.predict([[2,9,6]])
```

```python
reg1.predict([[12,10,10]])
```
