# DB Scan clustering

```python
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
```

```python
plt.scatter(X[:,0] ,X[:,1])
plt.show()
```

```python
db=DBSCAN(eps=3,min_samples=2)
db.fit(X)
db.labels_
```
