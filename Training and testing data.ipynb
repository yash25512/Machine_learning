# Split data into training and testing using sklearn 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv(r"C:\Users\namde\Downloads\carprices (1).csv")
df.head()
```

```python
plt.scatter(df['Mileage'],df['Sell Price($)'])
plt.xlabel('Mileage', fontsize=10)
plt.ylabel('Sell prices',fontsize=10)
plt.title('Mileage vs sell prices')
```

```python
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
plt.xlabel(' CarAGE', fontsize=10)
plt.ylabel('Sell prices',fontsize=10)
plt.title(' Car Age vs sell prices')
```

```python
X=df[['Mileage','Age(yrs)']]
y=df['Sell Price($)']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```
